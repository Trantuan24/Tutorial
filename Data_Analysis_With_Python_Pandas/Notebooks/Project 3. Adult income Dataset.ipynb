### 17. Replace Salary Values [ '<=50K', '>50K' ] With 0 and 1

```python
data.columns
```

```python
data['income'].unique()
```

```python
data['income'].value_counts()
```

```python
sns.countplot(x='income', data=data)
```

```python
def salary_data(s):
    if s == '<=50K':
        return 0
    else:
        return 1
```

```python
data.loc[:, 'enconded_salary'] = data['income'].apply(salary_data)
```
