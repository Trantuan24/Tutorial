### 20. Classify Movies Based on Ratings [Excellent, Good, and Average]

```python
data.columns
```

```python
def ratings(rating):
    if rating >= 7:
        return 'Excellent'
    elif rating >= 6: 
        return 'Good'
    else:
        return 'Average'
```

```python
data['rating_cat'] = data['Rating'].apply(ratings)
```

### 22. Find Unique Values From Genre 

```python
data['Genre']
```

```python
lst = []
for value in data['Genre']:
    lst.append(value.split(','))
```

```python
lst
```

```python
lst_item = []
for i in lst:
    for j in i:
        if j not in lst_item:
            lst_item.append(j)
len(lst_item)
```

```python
lst_item
```
